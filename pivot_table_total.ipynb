```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic_data.csv')
```

```python
df.head()
```

```python
df = df.drop(columns='PassengerId')
```

```python
df['Pclass'] = df.Pclass.replace({1: "First", 2: "Second", 3: "Third"})
```

```python
df.pivot_table("Survived", index=['Pclass'])
```

```python
df.pivot_table("Survived", index=['Pclass', "Sex"])
```

```python
df.pivot_table("Survived", index=['Pclass', "Sex"], aggfunc=np.mean)
```

```python
df.pivot_table("Fare", index=['Pclass', "Sex"], aggfunc=np.mean)
```

```python
df.pivot_table("Age", index=["Sex", "Survived", "Pclass"], columns = ["Embarked"], aggfunc=np.mean)
```

```python
df["Age"] = df["Age"].fillna(df.Age.mean())
```

```python
df.pivot_table("Age", index=["Sex", "Survived", "Embarked"], columns = ["Pclass"], aggfunc=np.mean)
```

```python
df.pivot_table("Age", index=["Sex", "Survived", "Embarked"], columns = ["Pclass"], aggfunc=np.mean, fill_value=0)
```

```python
df.pivot_table("Age", index=["Sex", "Survived", "Embarked"], columns = ["Pclass"], aggfunc=np.mean, fill_value=0)
```

```python
age = pd.cut(df["Age"], [0, 18, 80])
df.pivot_table("Survived", ["Sex", age], "Pclass")
```
