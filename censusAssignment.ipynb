```python
dict={}
for i in df.index:
    if df.loc[i]['STNAME'] not in dict:
        dict[df.loc[i]['STNAME']] = []
    dict[df.loc[i]['STNAME']].append(df.loc[i]['CENSUS2010POP'])
dict
```

```python
sm_dict = {}
for l in dict:
    sorted_l = sorted(dict[l], reverse=True)[:3]
    sm_dict[l] = sorted_l
sm_dict
    
```

```python
l_dict = {}
import numpy as np
for k in sm_dict:
    np_k = np.array(sm_dict[k])
    total = np_k.sum()
    l_dict[k] = total
l_dict
    
```

```python
result = sorted(l_dict, key = lambda key:l_dict[key], reverse=True)[:3]
result
```
